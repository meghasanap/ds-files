#### Standardization of Numerical columns

```python
df1
```

```python
# Creating an empty dataframe to store the standardize data
scaled_df1=pd.DataFrame(columns=['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'])
```

```python
scaled_df1
```

```python
# Calculating the standardize data (Z_score)
scaled_df1['Volume']=(df1['Volume']-df1['Volume'].mean())/df1['Volume'].std()
scaled_df1['Avg Price']=(df1['Avg Price']-df1['Avg Price'].mean())/df1['Avg Price'].std()
scaled_df1['Total Sales Value']=(df1['Total Sales Value']-df1['Total Sales Value'].mean())/df1['Total Sales Value'].std()
scaled_df1['Discount Rate (%)']=(df1['Discount Rate (%)']-df1['Discount Rate (%)'].mean())/df1['Discount Rate (%)'].std()
scaled_df1['Discount Amount']=(df1['Discount Amount']-df1['Discount Amount'].mean())/df1['Discount Amount'].std()
scaled_df1['Net Sales Value']=(df1['Net Sales Value']-df1['Net Sales Value'].mean())/df1['Net Sales Value'].std()
```

### Data Transfermation i:e converting the catagorical data into dummy variables 

```python
df2
```

```python
# Importing packages required for converting
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
# Using get_dummies function of pandas package
df_pandas = pd.get_dummies(df2, columns=['Date', 'Day'], drop_first=True)
df_pandas
```

```python
df_pandas = pd.get_dummies(df2, columns=['SKU'], drop_first=True)
df_pandas
```

```python
# Creating variable of functions to make is easy for using multiple times.
one_hot = make_pipeline(OneHotEncoder())
ord_enc = make_pipeline(OrdinalEncoder())
```

```python
# Using one hot encoder 
transformers=ColumnTransformer([('one_hot',one_hot,['BU'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
transformers=ColumnTransformer([('one_hot',one_hot,['City'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
# Using ordinal encoder
transformers=ColumnTransformer([('ord_enc',ord_enc,['Brand'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
transformers=ColumnTransformer([('ord_enc',ord_enc,['Model'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```
